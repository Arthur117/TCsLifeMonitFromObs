```python
### FUNCTIONS
def coriolis(lat):
    '''Latitude must be in degrees.'''
    Omega = 7.2921e-5                            # Earth rotation vector
    fcor  = 2 * Omega * np.sin(lat * np.pi / 180) # Coriolis parameter at 20° latitude and assuming it's constant 
    return fcor
```

```python
# Define scales
lat    = float(ds1.lat.isel(rad=0).mean(dim='theta', skipna=True))
# lat    = 5
fcor   = abs(coriolis(lat))                # s-1
Rmax   = float(ws1.rad[int(ws1.argmax())]) # m
Vmax   = float(ws1.max())                  # m/s
H      = 10000                             # m
Ro     = Vmax / (fcor * Rmax)
print('characteristic time --> {:.1f} h'.format(1 / fcor / 3600))
print('vertical velocity   --> {:.2f} m/s'.format(fcor * H))
print('radial velocity     --> {:.2f} m/s'.format(fcor * Rmax))
print('Rossby number       --> {:.0f}'.format(Ro))
```

```python
# Define non-dimensional (r, v)
r1 = ws1.rad / Rmax
v1 = ws1     / Vmax

V1 = xr.DataArray(
    data=v1,
    coords={'rad': r1}
)
```

```python
def v_k10(v0, t, Ro, lbda):
    v          = np.zeros_like(v0.rad)
    # Inner-core
    v_prop_in  = v0.interp({'rad': v0.rad * np.exp(t)})
    vin        = v_prop_in * np.exp((1 - lbda) * t) + v0.rad * (1 - np.exp((2 - lbda) * t)) / (Ro * (lbda - 2))
    # Near-core
    nr_term    = np.sqrt(1 + 2 * (np.log(v0.rad) + t))
    v_prop_nr  = v0.interp({'rad': nr_term})
    vnr        = nr_term * v_prop_nr * np.exp(-lbda * t) / v0.rad + (lbda * v0.rad ** 2 - 2 * v0.rad ** lbda) / (Ro * (lbda * v0.rad * (lbda - 2))) - np.exp(-lbda * t) / (Ro * (lbda * v0.rad))
    # Outer-core
    ou_term    = np.sqrt(v0.rad ** 2 + 2 * t)
    v_prop_ou  = v0.interp({'rad': ou_term})
    vou        = ou_term * v_prop_ou * np.exp(-lbda * t) / v0.rad + (1 - np.exp(-lbda * t)) / (Ro * (lbda * v0.rad))
    
    # Check each of the 3 pieces
    # plt.title('t={}'.format(t))
    # plt.plot(v0.rad, vin)
    # plt.plot(v0.rad, vnr)
    # plt.plot(v0.rad, vou)
    # plt.xlim(0, 3);plt.ylim(0, 3)
    # plt.grid();plt.show()
    
    # Declare final output
    bd              = np.exp(-t)
    v[v0.rad < bd]  = vin[v0.rad < bd]
    v[v0.rad >= bd] = vnr[v0.rad >= bd]
    v[v0.rad >= 1]  = vou[v0.rad >= 1]
    
    return v
```
