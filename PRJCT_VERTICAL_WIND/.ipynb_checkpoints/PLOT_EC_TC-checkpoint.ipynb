```python
def get_polar_theta(forward_az):
    '''INPUT: forward_az: 0° is North, 90° is East;
    OUTPUT: theta: 0° is East, 90° is North'''
    theta = (90 - forward_az) % 360 # 0° is North, 90° is East
    return theta

def add_polar_coords(d_EC, ds):
    # Interpolate on polar grid
    geod_wgs84        = CRS("epsg:4326").get_geod()
    forward_az, _, dr = geod_wgs84.inv(np.ones_like(d_EC.longitude) * float(ds.lon_ibt), np.ones_like(d_EC.latitude) * float(ds.lat_ibt), d_EC.longitude, ds_CFMR.latitude)
    forward_az        = forward_az - float(ds.storm_dir_ibt)
    th                = get_polar_theta(forward_az) # r is dr
    d_EC['r']         = xr.DataArray(data=dr, dims=['along_track'])
    d_EC['th']        = xr.DataArray(data=th, dims=['along_track'])
    # Relative r: positive North of the center, negative South of the center
    d_EC['rel_r']     = d_EC.r.copy(deep=True)
    d_EC['rel_r']     = xr.where(d_EC.latitude >= float(ds.lat_ibt), d_EC['rel_r'], -d_EC['rel_r'])
    return d_EC

def get_varY_from_varX(varY, varX, idxsX):
    ds2beInterped = xr.DataArray(
        data=np.array(ds_CFMR[varY]),
        coords={varX: np.array(ds_CFMR[varX])},
    )
    return ds2beInterped.interp(coords={varX: idxsX})
```

```python
# Custom cmap
from matplotlib.colors import LinearSegmentedColormap

# Parameters
threshold = 252  # Votre seuil de données
data_min  = 220  # La valeur minimale de vos données
data_max  = 280  # La valeur maximale de vos données

# Calcul du seuil normalisé (entre 0 et 1)
t_norm = (threshold - data_min) / (data_max - data_min)

# S'assurer que t_norm est dans la plage [0, 1]
t_norm = max(0.0, min(1.0, t_norm))

# --- 2. Obtenir les colormaps de base ---
cmap_turbo = plt.cm.get_cmap('turbo_r')
cmap_grays = plt.cm.get_cmap('Grays')

colors = []

num_steps_turbo = 100 # Nombre de pas pour échantillonner turbo
for i in range(num_steps_turbo + 1):
    pos = i / num_steps_turbo * t_norm # Position normalisée jusqu'au seuil
    colors.append((pos, cmap_turbo(pos / t_norm) if t_norm > 0 else cmap_turbo(0)))
    # La condition t_norm > 0 gère le cas où le seuil est 0

num_steps_grays = 100 # Nombre de pas pour échantillonner Grays
for i in range(num_steps_grays + 1):
    # La position normalisée de Grays va de 0 à 1 sur sa propre plage,
    # mais elle est mappée de t_norm à 1 sur la colormap combinée.
    pos_grays_internal = i / num_steps_grays
    pos_combined = t_norm + (1 - t_norm) * pos_grays_internal
    colors.append((pos_combined, cmap_grays(pos_grays_internal)))

# Sort the colors list by position to ensure correct interpolation
colors.sort(key=lambda x: x[0])

# Créer la colormap personnalisée
custom_cmap = LinearSegmentedColormap.from_list("turbo_grays_custom", colors)
```

```python
def coriolis(lat):
    '''Latitude must be in degrees.'''
    Omega = 7.2921e-5                             # Earth rotation vector
    fcor  = 2 * Omega * np.sin(lat * np.pi / 180) # Coriolis parameter assuming it's constant 
    return fcor
```

```python
dx = 5 # half_size of latitude box around TC center, in degrees

for oaf in tqdm(list_of_overpasses[151:]):
    try:
        ds = xr.open_dataset(glob.glob(PTHS['ear_TCs_dir'] + '*{}.nc'.format(oaf))[0])
    
        # Fig.
        fig        = plt.figure(figsize=(15, 14))
        gs         = GridSpec(1, 1, hspace=0.4, wspace=0.12, height_ratios=[1])
        title      = '{}, Lat = {:.1f}°N, Lon = {:.1f}°E, minDist2cen = {:.1f} km, Vmax = {:.1f} m/s, Rmax = {:.1f} km, t_since_LMI = {:.1f} hrs'.format(ds.atcf_id_ibt.item(), ds.lat_ibt, ds.lon_ibt, ds.closest_distance_to_track_center_ibt / 1000, ds.vmx_ibt, ds.rmx_ibt / 1000, ds.t_since_lmi_ibt / 3600)
        fig.suptitle(title, y=0.94, weight='bold')
        
        ##############
        ###  MCOP  ###
        ##############
        ax = fig.add_subplot(gs[0, :])
        # ax.text(0.95, 0.935, '(a)', transform=ax.transAxes, weight='bold')
    
        # Open
        ds_MCOP     = xr.open_dataset(glob.glob(PTHS['ear_dir'] + 'L2a/MSI_COP_2A/**/*{}.h5'.format(oaf), recursive=True)[0], group='/ScienceData')
        
        # Filter
        ds_MCOP_ftr = ds_MCOP.where((ds_MCOP.latitude >= float(ds.lat_ibt) - dx) & (ds_MCOP.latitude <= float(ds.lat_ibt) + dx))
        ds_MCOP_ftr = ds_MCOP_ftr.dropna(dim='across_track', subset=['longitude', 'latitude'], how='all').dropna(dim='along_track', subset=['longitude', 'latitude'], how='any')
    
        # Plot
        im = ax.pcolormesh(ds_MCOP_ftr.longitude, ds_MCOP_ftr.latitude, ds_MCOP_ftr.cloud_top_temperature, cmap=custom_cmap)
        
        # EC track
        ax.scatter(ds.longitude, ds.latitude, s=.1, c='r')
        # Center
        ax.scatter(ds.lon_ibt, ds.lat_ibt, marker='x', s=50, c='cyan', edgecolor='k', linewidth=2)
        # Wind radii
        num_pts = 1000
        geod = Geod(ellps='WGS84')
        azimuths = np.linspace(0, 360, 1000, endpoint=False)
        rmx_lons, rmx_lats, _ = geod.fwd(
            np.full(num_pts, float(ds.lon_ibt)), # Longitudes de départ (toutes identiques au centre)
            np.full(num_pts, float(ds.lat_ibt)), # Latitudes de départ (toutes identiques au centre)
            azimuths,                            # Azimuts pour chaque point du cercle
            np.full(num_pts, float(ds.rmx_ibt))  # Distances en mètres
        )
        r64_lons, r64_lats, _ = geod.fwd(
            np.full(num_pts, float(ds.lon_ibt)), # Longitudes de départ (toutes identiques au centre)
            np.full(num_pts, float(ds.lat_ibt)), # Latitudes de départ (toutes identiques au centre)
            azimuths,                            # Azimuts pour chaque point du cercle
            np.full(num_pts, float(ds.r64_ibt))  # Distances en mètres
        )
        rsby = ds.vmx_ibt / (coriolis(abs(ds.lat_ibt)) * ds.rmx_ibt)
        r0   = np.sqrt(2 * rsby) * ds.rmx_ibt
        r0_lons, r0_lats, _ = geod.fwd(
            np.full(num_pts, float(ds.lon_ibt)), # Longitudes de départ (toutes identiques au centre)
            np.full(num_pts, float(ds.lat_ibt)), # Latitudes de départ (toutes identiques au centre)
            azimuths,                            # Azimuts pour chaque point du cercle
            np.full(num_pts, float(r0))  # Distances en mètres
        )
        ax.plot(rmx_lons, rmx_lats, '-', color='k', linewidth=3)
        ax.plot(r64_lons, r64_lats, '-', color='k', linewidth=1)
        ax.plot(r0_lons,  r0_lats,  '-', color='k', linewidth=1)
    
        # Grid
        ax.set_xlim(min(ds_MCOP_ftr.longitude.min(), min(rmx_lons), min(r64_lons)), max(ds_MCOP_ftr.longitude.max(), max(rmx_lons), max(r64_lons)))
        ax.set_ylim(min(ds_MCOP_ftr.latitude.min(), min(rmx_lats), min(r64_lats)),  max(ds_MCOP_ftr.latitude.max(), max(rmx_lats), max(r64_lats)))
        ax.grid()
        # ax.set_aspect(0.5)
    
        # Colorbar
        divider = make_axes_locatable(ax)
        cax = divider.append_axes("right", size="3%", pad=0.1) # Create colorbar axis to the right
        fig.colorbar(im, cax=cax, label='Brightness temperature (K)')
    
        # Save
        plt.savefig(PTHS['sav_dir'] + 'MSI/MSI_{}'.format(oaf) + '.png', bbox_inches='tight');plt.clf()
        # plt.show()

        # Close

    except IndexError:
        print('No file for oaf = {}'.format(oaf))

    # Close
    ds.close()
```
