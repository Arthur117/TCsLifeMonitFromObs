### 1) GMF

```python
# Define paths of figs
p1 = PTHS['fgs_dir'] + 'GMF/2D_histo_density_plot_SMOS_vs_SAR_cyclobs_Dualpol_wind_speed_version=NEW.png'
p2 = PTHS['fgs_dir'] + 'GMF/2D_histo_density_plot_SMAP_vs_SAR_cyclobs_Dualpol_wind_speed_version=NEW.png'
p3 = PTHS['fgs_dir'] + 'GMF/scatter_3var_plot_SFMR_vs_SAR_cyclobs_Dualpol_wind_speed_version=NEW.png'
p4 = PTHS['fgs_dir'] + 'GMF/scatter_3var_plot_DROPSONDES_vs_SAR_cyclobs_Dualpol_wind_speed_version=NEW.png'
```

```python
# Créer une figure et une grille de sous-graphiques (2x2)
fig, axes = plt.subplots(2, 2, figsize=(10, 10))
axes = axes.flatten()  # Aplatir le tableau d'axes pour une itération facile

letters = ['(a)', '(b)', '(c)', '(d)']

for i, p in enumerate([p1, p2, p3, p4]):
    # Ouvrir l'image avec Pillow
    image = Image.open(p)

    # # Détecter et cropper les marges blanches
    # cropped_image = image.crop(image.getbbox())
    
    # Afficher l'image dans le sous-graphique approprié
    axes[i].imshow(image)
    
    # Ajouter la lettre en haut à droite
    axes[i].text(0.683, 0.13, letters[i], # 0.07, 0.91, letters[i],
                 transform=axes[i].transAxes,
                 # fontsize=20,
                 fontweight='bold')
                 # color='white',
                 # ha='right',
                 # va='top',
                 # bbox=dict(facecolor='black', alpha=0.5, edgecolor='none', boxstyle='round,pad=0.2'))
    
    # Supprimer les axes (graduation, étiquettes, etc.) pour une meilleure visualisation
    axes[i].axis('off')

# # Ajuster la mise en page pour éviter le chevauchement
# plt.tight_layout()
plt.subplots_adjust(wspace=-0.1, hspace=-0.2)

# # Save
# plt.savefig(PTHS['sav_dir'] + 'GMF_performances.png', bbox_inches='tight')
```
