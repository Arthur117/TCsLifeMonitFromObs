```python
### IMPORTANT PARAMETERS

# name and date of the tropical cyclone
#TC_name = str("'KATE'" + '_' + "'2003269N12322'")
TC_name = str("'SAM'" + '_' + "'2021262N10349'")

# Random walk scale
randw_scale = 0.1 # 0.05

# number of members
nb_members  = 100

# number of analogs
nb_analogs  = 100 #Â Don't touch

# variance of the observations (IBTrACS)
sigma2_obs  = 0.1 # 0.05

# variance of the SAR (for Rmax)
# sigma2_sar = 0.01

# index of the first unavailable observation
index_NaN = 60
```

```python
### GENERATE THE CATALOG, THE TRUE STATE, AND THE OBSERVATIONS

import os
files = os.listdir("../AnDA_data/IBTrACS_V3/.")
class catalog:
    IBTrACS_tmp = array(pandas.read_csv("../AnDA_data/IBTrACS_V3/" + files[0], usecols = ['Vmax', 'Rmax_CK22', 'R34', 'fcor', 'u_trans', 'v_trans', 'mean_ck_cd']))
    analogs = IBTrACS_tmp[0:-1,:]
    successors = IBTrACS_tmp[1:,:]
for i in arange(1, len(files)):
    if files[i] != TC_name:
        IBTrACS_tmp = array(pandas.read_csv("../AnDA_data/IBTrACS_V3/" + files[i], usecols = ['Vmax', 'Rmax_CK22', 'R34', 'fcor', 'u_trans', 'v_trans', 'mean_ck_cd']))
        catalog.analogs = concatenate([catalog.analogs, IBTrACS_tmp[0:-1,:]])
        catalog.successors = concatenate([catalog.successors, IBTrACS_tmp[1:,:]])
    else:
        class xt:
            time    = []
            values  = []
        IBTrACS_tmp = array(pandas.read_csv("../AnDA_data/IBTrACS_V3/" + files[i], usecols = ['Vmax', 'Rmax_CK22', 'R34', 'fcor', 'u_trans', 'v_trans', 'mean_ck_cd']))
        # IBTrACS_tmp = IBTrACS_tmp[:50]
        xt.time     = range(shape(IBTrACS_tmp)[0])
        xt.values   = IBTrACS_tmp
        class yo:
            time    = []
            values  = []
        IBTrACS_tmp = array(pandas.read_csv("../AnDA_data/IBTrACS_V3/" + files[i], usecols = ['Vmax', 'Rmax_CK22', 'R34', 'fcor', 'u_trans', 'v_trans', 'mean_ck_cd']))
        # IBTrACS_tmp = IBTrACS_tmp[:50]
        yo.time     = range(shape(IBTrACS_tmp)[0])
        yo.values   = IBTrACS_tmp
        yo.values[:, -1] = NaN # remove observations of Ck/Cd

# center and reduce the 6 columns
mean_IBTrACS = mean(catalog.analogs, axis=0)
std_IBTrACS  = std(catalog.analogs, axis=0)
for j in range(shape(catalog.analogs)[1]):
    catalog.analogs[:,j]    = (catalog.analogs[:,j] - mean_IBTrACS[j]) / std_IBTrACS[j]
    catalog.successors[:,j] = (catalog.successors[:,j] - mean_IBTrACS[j]) / std_IBTrACS[j]
    xt.values[:,j] = (xt.values[:,j] - mean_IBTrACS[j]) / std_IBTrACS[j]
    yo.values[:,j] = (yo.values[:,j] - mean_IBTrACS[j]) / std_IBTrACS[j]
```

```python
### Linear Regression to get Ck/Cd
from sklearn.linear_model import LinearRegression
Vmax     = xt.values[:,0] * std_IBTrACS[0] + mean_IBTrACS[0]
R17      = xt.values[:,2] * std_IBTrACS[2] + mean_IBTrACS[2]
fcor_abs = xt.values[:,3] * std_IBTrACS[3] + mean_IBTrACS[3]
ck_cd    = xt.values[:,6] * std_IBTrACS[6] + mean_IBTrACS[6]

# Vmax = Vmax[5:]
# ck_cd = ck_cd[5:]

lr = LinearRegression()
X  = concatenate((expand_dims(Vmax, axis=1), expand_dims(R17, axis=1), expand_dims(fcor_abs, axis=1)), axis=1)
lr.fit(X, ck_cd)
# plot(Vmax)
plot(lr.predict(X))
plot(ck_cd)
```

```python
### ANALOG DATA ASSIMILATION (dynamical model given by the catalog)

n = shape(xt.values)[1]

def get_rmax_ck22_augmented(Vmax, R17, fcor, Ck_Cd):
    '''Computes Rmax using CK22 model with the dependance on Ck_Cd
    INPUT:  Vmax (m/s), R17 (m), fcor (s-1), Ck_Cd ()
    OUTPUT: Rmax (km)
    '''
    
    # transform variables
    Vmax = Vmax * std_IBTrACS[0] + mean_IBTrACS[0]
    R17  = (R17 * std_IBTrACS[2] + mean_IBTrACS[2]) * 1000
    fcor = fcor * std_IBTrACS[3] + mean_IBTrACS[3]
    
    # force Ck_Cd to be positive using softplus function
    # Ck_Cd = log(1 + exp(Ck_Cd))
    
    # approximate Rmax
    M17      = R17 * 17.5 + 0.5 * fcor * (R17 ** 2)
    Mmax_M17 = (0.5 * Ck_Cd) ** (1 / (2 - Ck_Cd))     # Eq. 38 of ER11. CAVEAT: We must have R17 >> Rmax
    Mmax     = Mmax_M17 * M17
    # Mmax     = Mmax_M17 * M17 - (Vmax / 100) * fcor * (R17 ** 2)
    # Mmax     = Mmax_M17 * M17 - 0.1 * 0.5 * fcor * (R17 ** 2)
    # Mmax     = Mmax_M17 * M17 + (log(fcor) / 100) * R17 * 17.5
    # Mmax     = Mmax_M17 * M17 - 17.5 * R17 * (R17 / (fcor * Vmax))
    Rmax     = (Vmax / fcor) * (np.sqrt(1 + (2 * fcor * Mmax) / (Vmax ** 2)) - 1)
    
    return (Rmax / 1000 - mean_IBTrACS[1]) / std_IBTrACS[1] # return Rmax in km

def get_mean_ck_cd(Vmax, R17, fcor_abs):
    X  = concatenate((expand_dims(Vmax, axis=1), expand_dims(R17, axis=1), expand_dims(fcor_abs, axis=1)), axis=1)
    return lr.predict(X)
     

# parameters of the analog forecasting method
class AF:
    k                 = nb_analogs # number of analogs
    neighborhood      = np.ones((n,n)) # global analogs
    neighborhood[:,1] = 0 # do not use Rmax to find analogs ### ??? IMPORTANT ??? ###
    catalog           = catalog # catalog with analogs and successors
    regression        = 'local_linear' # chosen regression ('locally_constant', 'increment', 'local_linear')
    sampling          = 'gaussian' # chosen sampler ('gaussian', 'multinomial')
    
# parameters of the filtering method
class DA:
    method = 'AnEnKS'     # chosen method ('AnEnKF', 'AnEnKS', 'AnPF')
    N      = nb_members   # number of members (AnEnKF/AnEnKS) or particles (AnPF)
    xb     = xt.values[0, :]; 
    xb[-1] = (0.5 - mean_IBTrACS[-1]) / std_IBTrACS[-1]
    B      = np.eye(n)
    H      = np.eye(n)
    R      = sigma2_obs * np.eye(n)
    # R[1, 1]= 0.01
    # R[1,1] = sigma2_sar ### VARIANCE OF THE SAR MEASUREMENTS ###
    def m(x):
        x_out = zeros((2, DA.N, n))
        x_out[:, :, :] = AnDA_analog_forecasting(x, AF)
        mins           = ones_like(x_out[0, :, 6]) * 0.01
        ck_cd_pos      = where(mins < x_out[0, :, 6], x_out[0, :, 6], 0.01)
        rmax_ck_cd     = get_rmax_ck22_augmented(x_out[0, :, 0], x_out[0, :, 2], x_out[0, :, 3], ck_cd_pos)
        #Â Mean between Rmax_ck22 and Rmax_ck22_with_ck_cd
        # x_out[0, :, 1] = (rmax_ck_cd + x_out[0, :, 1]) / 2.
        x_out[0, :, 1] = rmax_ck_cd
        x_out[1, :, 1] = mean(x_out[0, :, 1])
        
        return x_out

# run the analog data assimilation
x_hat_analog = AnDA_data_assimilation(yo, DA)
```

```python
# compute weighted quantiles
x_hat_analog_quantiles = zeros([shape(x_hat_analog.part)[0], 7, n])
for t in range(shape(x_hat_analog.part)[0]):
    for j in range(n):
        wq = DescrStatsW(data=x_hat_analog.part[t,:,j], weights=x_hat_analog.weights[t,:])
        x_hat_analog_quantiles[t,:,j] = wq.quantile(probs=np.array([0.025, 0.05, 0.25, 0.5, 0.75, 0.95, 0.975]), return_pandas=False)
```

```python
from scipy import optimize

def ER11_model_ratio(Ck_Cd):
    return (0.5 * Ck_Cd) ** (1 / (2 - Ck_Cd))

def ER11_subtraction(Ck_Cd, ER11_SAR_ratio):
    return ER11_model_ratio(Ck_Cd) - ER11_SAR_ratio

def fit_Ck_Cd_eq38(ER11_SAR_ratio):
    '''Hypothesis of ER11: Rmax >> R17'''
    sol = optimize.root(
        ER11_subtraction,
        x0=np.ones_like(ER11_SAR_ratio) * 0.5, #Â first variable: Ck_Cd
        args=ER11_SAR_ratio,                   #Â fixed parameters
        method='lm',                           #Â hybr method doesn't work for ER11_SAR_ratio < 0.3
        tol=1e-6
    )
    return sol.x

# Compute Ck/Cd with ER11 and Rmax_CK22
Vmax     = xt.values[:,0] * std_IBTrACS[0] + mean_IBTrACS[0]
Rmax     = xt.values[:,1] * std_IBTrACS[1] + mean_IBTrACS[1]
R17      = xt.values[:,2] * std_IBTrACS[2] + mean_IBTrACS[2]
fcor_abs = xt.values[:,3] * std_IBTrACS[3] + mean_IBTrACS[3]

ER11_SAR_ratio = Rmax * 1000 * Vmax / (R17 * 1000 * 17.5 + 0.5 * fcor_abs * ((R17 * 1000) ** 2))
Ck_Cd2         = fit_Ck_Cd_eq38(ER11_SAR_ratio)

# scatter(ER11_model_ratio(Ck_Cd2), ER11_SAR_ratio)
```
